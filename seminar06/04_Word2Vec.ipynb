## Word2Vec

Будем обучать word2vec из пакета gensim на корпусе opencorpora.<br/>
Этот корпус небольшой, поэтому выдающихся результатов ждать не стоит, но зато обучение будет происходить быстро.

```python
import gensim
from gensim.models.word2vec import LineSentence, Word2Vec

sentences = LineSentence('opencorpora_for_word2vec.txt')
```

```python
len(list(sentences))
```

```python
list(sentences)[3]
```

```python
model = Word2Vec(sentences, size=300, window=5, min_count=5, workers=4, iter=20)
model.init_sims(replace=True)
```

```python
for w, sim in model.wv.most_similar(positive=['Google']):
    print(w, sim)
```

```python
for w, sim in model.wv.most_similar(positive=['оператор']):
    print(w, sim)
```

```python
for w, sim in model.wv.most_similar(positive=['мальчик', 'женщина'], negative=['мужчина']):
    print(w, sim)
```

```python
for w, sim in model.wv.most_similar(positive=['ходить']):
    print(w, sim)
```

```python
for w, sim in model.wv.most_similar(positive=[u'брат', u'жена'], negative=[u'муж']):
    print(w, sim)
```

```python
print(model.wv.doesnt_match("книга журнал машина".split()))
```

```python
model.wv.similarity('книга', 'телефон')
```
